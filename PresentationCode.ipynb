```python
# smoker or non-smoker
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(data.smoker.drop_duplicates()) 
data.smoker = le.transform(data.smoker)
#region
le.fit(data.region.drop_duplicates()) 
data.region = le.transform(data.region)
```

```python
# After seeing correlation, we will now study the distribution of variables in the dataset
# We create subplots for a referential distribution study as compared to plotting each histogram one by one
fig, axes = pl.subplots(nrows=2, ncols=2, figsize=(12, 10))
data.plot(kind="hist", y="age", bins=70, color="b", ax=axes[0][0])
data.plot(kind="hist", y="bmi", bins=200, color="r", ax=axes[0][1])
data.plot(kind="hist", y="children", bins=5, color="g", ax=axes[1][0])
data.plot(kind="hist", y="charges", bins=200, color="orange", ax=axes[1][1])
pl.show()
```

```python
#We will now count the frequency of smokers and non-smokers where smoker = 1 and non-smoker=0
sns.catplot(x="smoker", kind="count", palette="Set1", data=data)
```

```python
# count the number of males and females where female=1 and male =0
sns.catplot(x="sex", kind="count", palette="Set1", data=data)
```

```python
fig, axes = pl.subplots(nrows=1, ncols=3, figsize=(18, 5))
data.plot(kind='scatter', x='age', y='charges', alpha=0.5, color='green', ax=axes[0], title="Age vs. Charges")
data.plot(kind='scatter', x='bmi', y='charges', alpha=0.5, color='red', ax=axes[1], title="BMI vs. Charges")
data.plot(kind='scatter', x='smoker', y='charges', alpha=0.5, color='blue', ax=axes[2], title="Smoker vs. Charges")
pl.show()
```

```python

# The scatter plots show that as Age increases charges also increase
#As BMI increases Charges increase
# A smoker has higher charges as compared to a non-smoker
# we will start by seeing distribution of charges for smokers and non- smokers
f= pl.figure(figsize=(12,5))
ax=f.add_subplot(121)
sns.distplot(data[(data.smoker == 1)]["charges"],color='c',ax=ax)
ax.set_title('Distribution of charges for smokers')

ax=f.add_subplot(122)
sns.distplot(data[(data.smoker == 0)]['charges'],color='b',ax=ax)
ax.set_title('Distribution of charges for non-smokers')
```

```python
# Count the number of male and female smokers . Our dataset has male=0 and female=1
sns.catplot(x="smoker", kind="count",hue = 'sex', palette="pink", data=data)
```

```python
# The count of number of non-smokers is higher than smokers. Also, we see that male smokers are more as compared to female smokers
#Thus, we can assume at this point that male members have higher charges as compared to females.
pl.figure(figsize=(12,5))
pl.title("Box plot for charges of women")
sns.boxplot(y="smoker", x="charges", data =  data[(data.sex == 1)] , orient="h", palette = 'magma')
```

```python
pl.figure(figsize=(12,5))
pl.title("Box plot for charges of men")
sns.boxplot(y="smoker", x="charges", data =  data[(data.sex == 0)] , orient="h", palette = 'rainbow')
```

```python
#create a new column for obesity where bmi>=30
data['obese'] = np.where(data['bmi']>=30, 1, 0)
```

```python
data.head()
```

```python
model = ols("charges ~ obese*smoker", data=data)
results = model.fit()
results.summary()
```

```python
results.predict({'obese':1,'smoker':1})
```

```python
# From the model we can see that obesity increases health care cost by $865.67 and smoking increases the cost by $13,390
#However, the combined effect of obesity and smoker is more prominent with a increase in cost by $19,330
```
